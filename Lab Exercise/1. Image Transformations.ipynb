### Question : 
- Take any colour image. Ensure that the resolution of the image is 512x512. 
- Convert this into a gray scale image. 
- Write  python codes to perform any 5 transformations on the image. (Example: obtain the negative of the image.) 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
# using PIL
img_pil = Image.open('Images/dog.jpg') 

# using cv2
img_cv2 = cv2.imread('Images/dog.jpg')
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title('PIL Image')
plt.imshow(img_pil)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('CV2 Image')
img_cv2 = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB) # convert to RGB (it was in BGR)
plt.imshow(img_cv2)
plt.axis('off')
```

```python
img_pil.size, img_cv2.shape
```

```python
# convert to 512 x 512
img_pil = img_pil.resize((512,512))
img_cv2 = cv2.resize(img_cv2, (512,512))
img_pil.size, img_cv2.shape
```

```python
# convert to grayscale
img_pil_gray = img_pil.convert('L')
img_cv2_gray = cv2.cvtColor(img_cv2, cv2.COLOR_RGB2GRAY)
img_pil_gray.size, img_cv2_gray.shape
```

```python
# print the image
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_pil_gray, cmap='gray')
plt.title('PIL')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_cv2_gray, cmap='gray')
plt.title('cv2')
plt.axis('off')
```

`    Transformation 1: Horizontal Flip`

```python
img_pil_flip = img_pil.transpose(Image.FLIP_LEFT_RIGHT)
img_cv2_flip = cv2.flip(img_cv2, 1)
```

```python
plt.figure(figsize=(15,5))

# original image
plt.subplot(1,3,1)
plt.imshow(img_pil)
plt.title('Original Image')
plt.axis('off')

# PIL flipped image
plt.subplot(1,3,2)
plt.imshow(img_pil_flip)
plt.title('PIL Flipped Image')
plt.axis('off')

# OpenCV flipped image
plt.subplot(1,3,3)
plt.imshow(img_cv2_flip)
plt.title('OpenCV Flipped Image')
plt.axis('off')

plt.show()
```

`Transformation 2: Rotation`

```python
img_pil_rot = img_pil.rotate(180)
img_cv2_rot = cv2.rotate(img_cv2, cv2.ROTATE_180)

# print the image
plt.figure(figsize=(15,5))

# original image
plt.subplot(1,3,1)
plt.imshow(img_pil)
plt.title('Original Image')
plt.axis('off')

# rotated image
plt.subplot(1,3,2)
plt.imshow(img_pil_rot)
plt.title('PIL Rotated Image')
plt.axis('off')

# OpenCV rotated image
plt.subplot(1,3,3)
plt.imshow(img_cv2_rot)
plt.title('OpenCV Rotated Image')
plt.axis('off')

plt.show()
```

`Transformation 3: Cropping the image`

```python
img_pil_crop = img_pil.crop((100,100,400,400))
img_cv2_crop = img_cv2[100:400,100:400]
```

```python
plt.figure(figsize=(15,5))

# original image
plt.subplot(1,3,1)
plt.imshow(img_pil)
plt.title('Original Image')
plt.axis('off')

# PIL cropped image
plt.subplot(1,3,2)
plt.imshow(img_pil_crop)
plt.title('PIL Cropped Image')
plt.axis('off')

# OpenCV cropped image
plt.subplot(1,3,3)
plt.imshow(img_cv2_crop)
plt.title('OpenCV Cropped Image')
plt.axis('off')

plt.show()
```

` Transformation 4: Image Translation`

```python
M = np.float32([[1,0,200],[0,1,50]])
translated_img = cv2.warpAffine(img_cv2, M, (img_cv2.shape[1], img_cv2.shape[0]))

# print the image
plt.figure(figsize=(10,5))

# original image
plt.subplot(1,2,1)
plt.imshow(img_cv2)
plt.title('Original Image')
plt.axis('off')

# translated image
plt.subplot(1,2,2)
plt.imshow(translated_img)
plt.title('Translated Image')
plt.axis('off')

plt.show()
```

`Transforamtion 5: Brightness Adjustment`

```python
# Transforamtion 5: Brightness Adjustment
adjusted_img = cv2.convertScaleAbs(img_cv2, alpha=1.5, beta=1)

# print the image
plt.figure(figsize=(10,5))

# original image
plt.subplot(1,2,1)
plt.imshow(img_cv2)
plt.title('Original Image')
plt.axis('off')

# adjusted image
plt.subplot(1,2,2)
plt.imshow(adjusted_img)
plt.title('Adjusted Image')
plt.axis('off')

plt.show()
```

`Transformation 6: Negative of the image`

```python
# obtain negative image
negative_img = cv2.bitwise_not(img_cv2)

# print the image
plt.figure(figsize=(10,5))

# original image
plt.subplot(1,2,1)
plt.imshow(img_cv2)
plt.title('Original Image')
plt.axis('off')

# negative image
plt.subplot(1,2,2)
plt.imshow(negative_img)
plt.title('Negative Image')
plt.axis('off')

plt.show()
```
