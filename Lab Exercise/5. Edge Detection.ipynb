### Question :
Given the input, apply blur, edge detection and other image processing approach to get the Expected output image

### Import packages and libraries and read the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image_new.jpeg")
img_ex = cv2.imread("Expected Output.jpg")
```

```python
print(img.shape)
print(img_ex.shape)
```

### Plot the images

```python
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Input')

plt.subplot(1, 2, 2)
plt.imshow(img_ex)
plt.title('Expected Output')

plt.show()
```

### HSV Conversion

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv_img)
```

### Creating a mask

```python
lower_white_limit = np.array([0, 0, 200]) 
upper_white_limit = np.array([180, 50, 255])

mask = cv2.inRange(hsv_img, lower_white_limit, upper_white_limit)
```

```python
plt.imshow(mask,cmap='gray')
```

### Overlap Image

```python
inverse_mask = 255 - mask
final_image = cv2.bitwise_and(img, img, mask=inverse_mask)
plt.imshow(final_image)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.imshow(final_image)
plt.title('Output Image')

plt.subplot(1, 2, 2)
plt.imshow(img_ex)
plt.title('Expected Output')

plt.show()
```
