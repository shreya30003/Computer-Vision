#### Question:

Extract SIFT or SURF features from an image and visualize them. 

`Import the required libraries`

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

`Function to display the images`

```python
def display(img,name):
    plt.title(name)
    plt.imshow(img,cmap='gray') 
    plt.axis('off')
```

`Read the image`

```python
image = cv2.imread('Test-slanted.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

display(gray,'Gray Image')
```

`Image Preprocessing`

```python
# Initialize the SURF detector
surf = cv2.xfeatures2d.SURF_create()

# Detect SURF features, with no masks
keypoints, descriptors = surf.detectAndCompute(gray, None)

# Draw the keypoints
img = cv2.drawKeypoints(image, keypoints, None)

# Display the image with SURF features
plt.imshow(img)
plt.show()
```
