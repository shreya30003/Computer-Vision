`Import the required libraries`

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

`Function to display the images`

```python
def display(img,name):
    plt.title(name)
    plt.imshow(img,cmap='gray') 
    plt.axis('off')
```

`Read the image`

```python
image = cv2.imread('crack.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(gray,'Gray Image')
```

`Image Processing`

```python
blur = cv2.GaussianBlur(gray,(3,3),2)
display(blur,'Blurred Image')
```

```python
# Increase brightness and contrast
alpha = 2
beta = 0.5
bright = cv2.convertScaleAbs(blur, alpha=alpha, beta=beta)
display(bright,'Bright Image')
```

```python
# Edge Detection
edges = cv2.Canny(bright, 150, 200)
display(edges,'Edges')
```

```python
_,  bin_img = cv2.threshold(bright, 127, 255, cv2.THRESH_BINARY) 
display(bin_img,'Binary Image')
```

```python
_,  bin_inv_img = cv2.threshold(bright, 127, 255, cv2.THRESH_BINARY_INV) 
display(bin_inv_img,'Binary Inverted Image')
```

```python
_,  bin_img_otsu = cv2.threshold(bright, 127, 255, cv2.THRESH_OTSU) 
display(bin_img_otsu,'Binary Otsu Image')
```

```python
_,  bin_img_otsu_bin = cv2.threshold(gray, 140, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) 
display(bin_img_otsu_bin,'Binary Otsu Image')
```

```python
_,  bin_inv_img_otsu_bin = cv2.threshold(bright, 127, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU) 
display(bin_inv_img_otsu_bin,'Binary Inverted Otsu Image')
```
