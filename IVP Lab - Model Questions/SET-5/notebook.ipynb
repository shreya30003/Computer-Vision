`Import the required libraries`

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

`Function to display the images`

```python
def display(img,name):
    plt.title(name)
    plt.imshow(img,cmap='gray') 
    plt.axis('off')
```

`Read the image`

```python
image = cv2.imread('Art.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(gray,'Original Image')
```

`Segregating Foreground and Background Images`

```python
ret1, thresh1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
display(thresh1,'Global Thresholding')
```

```python
thresh2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 14)
thresh3 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 21)
'''
cv2.adaptiveThreshold(src, maxValue, adaptiveMethod, thresholdType, blockSize, C)
src: Input gray scale image
maxValue: Maximum intensity value which is assigned to output
adaptiveMethod: Adaptive method to calculate threshold
thresholdType: Type of thresholding -> cv2.THRESH_BINARY or cv2.THRESH_BINARY_INV
blockSize: Size of a pixel neighborhood that is used to calculate a threshold value for the pixel
C: Constant subtracted from the mean or weighted mean
'''
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121), display(thresh2,'Adaptive Mean Thresholding')
plt.subplot(122), display(thresh3,'Adaptive Gaussian Thresholding')
plt.show()
```
