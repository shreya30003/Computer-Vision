`Methods to measure image quality : `
- PSNR
- SSIM
- LPIPS

`PSNR`

Peak Signal-to-noise ratio : \
Ratio between maximum possible power of an image and the power of the corrupting noise which affects the quality of the image
- PSNR = 10 log(R^2/ MSE)
- R = 255

````python
def PSNR(original, distorted): 
    mse = np.mean((original - distorted)**2)
    if mse == 0: # no noise
        return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse))
    return psnr
````

`SSIM : Structural Similarity Index`

Asseses Structural Similarity between two images. Considers luminance, contrast and structure. Gives a score between -1 (dissimilar) and 1 (identical)

````python
def SSIM(original, distorted):
    img1_gray = original
    img2_gray = distorted
    ssim = metrics.structural_similarity(img1_gray, img2_gray, full = True)
    return round(ssim[0], 2)
````

` LPIPS : Learned Perpetual Image Patch Similarity`

Computes the similarity between the activations of two image patched for some pre-defined network.

````python
def LPIPS(original, distorted):
    # Convert grayscale images to RGB
    if len(original.shape) == 2:
        original = cv2.cvtColor(original, cv2.COLOR_GRAY2RGB)
    if len(distorted.shape) == 2:
        distorted = cv2.cvtColor(distorted, cv2.COLOR_GRAY2RGB)

    img1 = torch.from_numpy(np.array(original)).unsqueeze(0).float()
    img2 = torch.from_numpy(np.array(distorted)).unsqueeze(0).float()
    
    img1 = img1.permute(0, 3, 1, 2).float()/127.5 -1
    img2 = img2.permute(0, 3, 1, 2).float()/127.5 -1

    lpips = LearnedPerceptualImagePatchSimilarity(net_type='squeeze')
    return lpips(img1,img2).item()
````

`MAP Estimation`

````python
def map_restore(og_img, kernel, nlevel, iters):
    temp_img = og_img.copy()
    for _ in range(iters):
        blur_estimate = convolve2d(temp_img, kernel, mode='same', boundary='wrap') 
        # This line estimates the blur in the image by convolving temp_img with a blur kernel. 
        # The 'same' mode means the output will be the same size as temp_img, and the 'wrap' boundary option means that the input is assumed to be periodic.
        
        data_likelihood = np.exp(-0.5 * ((og_img - blur_estimate) / nlevel) ** 2) 
        # This line calculates the data likelihood, which is the probability of the observed image given the estimated image and the noise level.
        
        prior_likelihood = np.ones_like(temp_img)  
        # Uniform prior -> we assume all possible images are equally likely, sets up a matrix of ones with the same shape as temp_img
        
        posterior = data_likelihood * prior_likelihood
        temp_img = posterior / np.sum(posterior) # normalize the posterior
    
    return 1 - temp_img # returns the negative of the estimated image. This is done to invert the colors of the image
````

````python
kernel = np.ones((5, 5)) / 25
````

````python
map_estimation_01 = map_restore(noisy_image_01, kernel, 100, 100)
map_estimation_02 = map_restore(noisy_image_02, kernel, 100, 100)
````

````python
plt.figure(figsize=(10,10))
plt.subplot(221), display(noisy_image_01,'Noisy Image') 
plt.subplot(222), display(map_estimation_01,'MAP Estimated Image')
plt.subplot(223), display(noisy_image_01,'Noisy Image')
plt.subplot(224), display(map_estimation_02,'MAP Estimated Image')
plt.show()
````

````python
print('PSNR of Gray Image and Noisy Image : ', PSNR(gray_img, noisy_image_01))
print("PSNR of Gray Image and MAP Estimated Image : ", PSNR(gray_img, map_estimation_01))
print('\n')
print('PSNR of Gray Image and Noisy Image : ', PSNR(gray_img, noisy_image_02))
print("PSNR of Gray Image and MAP Estimated Image : ", PSNR(gray_img, map_estimation_02))
````

`CNN`

````python
noisy_image_01.shape, noisy_image_02.shape, gray_img.shape
````

````python
model = Sequential([
    Input(shape=(1001, 1500, 1)),
    Conv2D(32, (3, 3), activation='relu'),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1001 * 1500, activation='linear')
])
````

````python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
````

````python
noisy_image_01_reshaped = np.reshape(noisy_image_01, (-1, 1001, 1500, 1))
gray_img_reshaped = gray_img.reshape(1, 1001 * 1500)

model.fit(noisy_image_01_reshaped, gray_img_reshaped, epochs=2)

# Predict/denoise noisy_image_01
denoised_image_01 = model.predict(noisy_image_01_reshaped)
````

````python
model.fit(noisy_image_02, gray_img, epochs=2)

# Predict/denoise noisy_image_02
denoised_image_02 = model.predict(noisy_image_02)
````

````python
print('PSNR of Gray Image and Noisy Image : ', PSNR(gray_img, noisy_image_01))
print("PSNR of Gray Image and CNN Image : ", PSNR(gray_img, denoised_image_01))
print('\n')
print('PSNR of Gray Image and Noisy Image : ', PSNR(gray_img, noisy_image_02))
print("PSNR of Gray Image and CNN Image : ", PSNR(gray_img, denoised_image_02))
````

### Question 2 :

Apply smoothing filters in spatial domain

Smoothing filters in the spatial domain are used to reduce noise and detail in an image. They work by replacing each pixel's value with the average value of the pixel values in its neighborhood. Here are a few common smoothing filters:

1. **Box (or Mean) Filter**: This filter replaces each pixel's value with the average value of its neighbors, including itself. This is equivalent to convolving the image with a kernel of equal coefficients.

    ```python
    blur = cv2.blur(img, (5, 5))  # Kernel size is 5x5
    ```

2. **Gaussian Filter**: This filter uses a Gaussian kernel instead of a box filter. This results in a smoother and more natural effect.

    ```python
    blur = cv2.GaussianBlur(img, (5, 5), 0)  # Kernel size is 5x5
    ```

3. **Median Filter**: This filter replaces each pixel's value with the median value of its neighbors. It's very effective at removing salt-and-pepper noise.

    ```python
    median = cv2.medianBlur(img, 5)  # Kernel size is 5
    ```

4. **Bilateral Filter**: This filter is more complex and takes longer to compute, but it has the advantage of preserving edges while reducing noise.

    ```python
    blur = cv2.bilateralFilter(img, 9, 75, 75)  # Diameter, SigmaColor, SigmaSpace
    ```
